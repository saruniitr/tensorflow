# Trigonometric plots using Tensorflow

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(-6, 6, 100)
```

```python
X = tf.placeholder(dtype=tf.float32, shape=x.size)
siny = tf.sin(X)
cosy = tf.cos(X)
```

```python
init = tf.global_variables_initializer()
```

```python
with tf.Session() as ssn:
    ssn.run(init);
    sin_y = ssn.run(siny, feed_dict={X: x});
    cos_y = ssn.run(cosy, feed_dict={X: x});
```

```python
plt.subplot(2,1,1)
plt.plot(x, sin_y, 'b')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine wave')
plt.grid(True)

# now select second plot in the 2x1 array
plt.subplot(2,1,2)
plt.plot(x, cos_y, 'r')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.title('Cosine wave')
plt.grid(True)
```
